```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.metrics as sk
import scipy.stats
from scipy.optimize import curve_fit
from RegscorePy import *
from IPython.core.debugger import set_trace
import lmfit
##### This script is to analyze city-by-city income distributions
```

```python
data = pd.read_csv("CTScalingFiles/ct.distr.csv")
idmappings = pd.read_excel("CTScalingFiles/cbsaidmappings.xls").set_index("CBSA Code")

# maps MSA id to census tract incomes
income_lists = {}
# maps MSA id to census tract population
pops = {}
fullpops = {}
cts = set()
i = 0
# populations of whole cities (TODO: Cross-check with other data)
city_pops = {}

metro_ids = set()
micro_ids = set()
for _,ct in data.iterrows():
    i +=1
    msa = ct['msa']
    ct_id = ct["ct"]
    
    #NOTE on data: "meandollars is total ct income / # workers... "
    ct_inc = ct["ct.meandollars"]
    pop = ct["ct.totalworkers"]
    fullpop = ct["ct.population"]
    if (ct["ct"],msa) in cts:
        print("WARNING! Found Double census tract id: "+str(ct["ct"]))
        set_trace()
    cts.add((ct["ct"],msa))
    
    if msa not in city_pops:
        city_pops[msa] = ct["pop.sum"]
    
    if msa not in income_lists or msa not in pops:
        income_lists[msa] = []
        pops[msa] = []
        fullpops[msa] = []
    income_lists[msa].append(ct_inc)
    pops[msa].append(pop)
    fullpops[msa].append(fullpop)
    msa = str(msa)
    metro_micro_lbl = idmappings["Metropolitan/Micropolitan Statistical Area"][msa]
    if not isinstance(metro_micro_lbl,str):
        metro_micro_lbl = list(metro_micro_lbl)[0]
    metro = "Metro" in metro_micro_lbl
    if metro:
        metro_ids.add(msa)
    else:
        micro_ids.add(msa)

```

```python
# just to test how many census tracts there are in different cities

n_cts = {}
for _,ct in data.iterrows():
    msa = ct['msa']
    if str(msa) in metro_ids:
        ct_id = ct["ct"]
        if msa not in n_cts:
            n_cts[msa] = 0
        n_cts[msa] += 1
    
n_cts_list = []
for msa in income_lists:
    if str(msa) in metro_ids:
        n_cts_list.append(n_cts[msa])
    
plt.figure()
plt.hist(n_cts_list,bins = 150)
plt.show()
```

```python
def sigmoid(x, k=3.):
    return 1/(1+np.exp(-k*x))


# plt.figure()
# x = np.linspace(0,20,100)
# plt.plot(x,sigmoid(x-10.))
# plt.show()

def piecewise_model(x,kappa,g_alpha,g_beta,g_norm,p_b,p_norm):
    #print(kappa)
    
    # x is a list of values, must return list of outputs
    
    # gamma distribution
    
    gf = scipy.stats.gamma(float(g_alpha),1./float(g_beta))
    # TODO: Test with coefficient parameter
    # pareto distribution
    pf = scipy.stats.pareto(float(p_b))
    ret = []
    for inc in x:
        # gamma distribution for x < cutoff kappa
        gval = g_norm*gf.pdf(inc)
        # pareto distribution for x > cutoff kappa
        pval = p_norm*pf.pdf(inc)
        # combine with sigmoid
        combined = gval*sigmoid(kappa-inc) + pval*sigmoid(inc-kappa)
        #print("combined",combined)
        ret.append(combined)
    #print("ret",ret)
            
        
            
    #print("input:",x)
#     print("output:",ret)
    return ret

def piecewise_residual(params,x,data):
    model = piecewise_model(x,params['g_alpha'],params['g_beta'],params['p_b'],params['kappa'],params['p_s'])
    return np.array(data) - np.array(model)


modelType = "piecewise"

# kappa = cutoff value between distributions
modelParams = {
    "piecewise":["kappa","g_alpha","g_beta","p_b","p_s"]
}

mini = {}
fullFits = {}
fullModels = {} 
fullParams = {}

params = []
city_pop_list = []

for msa in hists:
    
    # skip over anything not an MSA
    if str(msa) in micro_ids:
        continue
    num_people,incomes,_ = hists[msa]
    incomes = incomes[:-1]
    normed_np = [float(i)/sum(num_people) for i in num_people]
    num_people = normed_np
    #num_people_cleaned = []
    city_pop = city_pops[msa]
    city_pop_list.append(city_pop)
    
    #only look at large cities, with better data
    if city_pop <1000000:
        continue
    
    # remove bins with no population
    cleaned_np = []
    cleaned_inc = []
    for (ctpop,ctinc) in zip(num_people,incomes):
        if not ctpop == 0:
            cleaned_np.append(ctpop)
            cleaned_inc.append(ctinc)
    
            city_pop_all.append(np.log(city_pop))
        
    # rescale, x becomes $10,000
    cleaned_inc = list(np.array(cleaned_inc)*1./10000.)
    #fullParams[msa] = lmfit.Parameters()
    #fullParams[msa].add_many(('kappa',7.5,True,1.,30.),
    #                    ('g_alpha',3.,True,0.),
    #                    ('g_beta',1.,True,0.),
    #                    ('p_b',2.,True,0.,3.5),
    #                         #TODO: Check
    #                    ('p_s',10.,True))
    #
    #mini[msa] = lmfit.Minimizer(piecewise_residual,fullParams[msa],fcn_args=(cleaned_inc, cleaned_np))
    #
    #fullFits[msa] = mini[msa].minimize(params=fullParams[msa])#mini.minimize(params=fullParams)
    #fullModels[msa] = lmfit.Model(piecewise_model)
    #fullFits[msa].params.pretty_print()
#   #  fullParams[msa] = fullFits[msa].params
#   #  fullParams[msa].pretty_print()
    #
    ##TODO: plot
    #fullParams[msa] = fullFits[msa].params
    guess = [10.,3.,1.,0.5,2.,2.]
    bounds = ([1.,0.,0.,0.,-np.inf,0.],[30.,np.inf,np.inf,3.5,np.inf,50.])
    
    #TODO: Some cases of overfitting***!!! Two bumps, between two fns
    #TODO: Fit with different number of bins above; maybe do have to do some extrapolation?
    msaname = idmappings["CBSA Title"][str(msa)]
    if not isinstance(msaname,str):
        msaname = list(msaname)[0]
    plt.figure()
    plt.plot(cleaned_inc,cleaned_np,'o')
    plt.show()
    print("fitting: ",msaname)
    coeff, var = curve_fit(piecewise_model, cleaned_inc, cleaned_np,p0=guess,bounds=bounds)
    x = np.linspace(min(cleaned_inc),max(cleaned_inc),500)
    #plt.plot(x,piecewise_model(x,fullParams[msa]['g_alpha'],fullParams[msa]['g_beta'],fullParams[msa]['p_b'],fullParams[msa]['kappa'],fullParams[msa]['p_s']))
    plt.plot(cleaned_inc,cleaned_np,'o')
    plt.plot(x,piecewise_model(x,*coeff))
    print(coeff)
    plt.xlabel("income")
    plt.ylabel("probability")
    params.append(coeff)
    
    
    "showing plot"
    plt.show()
    

"""TODO: 
- Look into where data comes from - what exactly does it include?
- Josh's model
- compare data of ratio top/bottom to what we have here
"""

print("finished")



    
```

# Reproduce Decile Scaling result

```python
# maps int (1-10) to pair (list of income_pp,list of city size)
deciles = {}
for i in range(1,11):
    deciles[i] = ([],[])

for msa in income_lists:
    if str(msa) in metro_ids:
        income_list = income_lists[msa]
        pop_list = fullpops[msa]
        city_pop = city_pops[msa]

        #split into deciles
        sum_pop = sum(pop_list)
        pop_cutoffs = [sum_pop*float(i)/10. for i in range(1,11)]
        cumulative_pop = 0.
        current_decile = 1
        # summed population and income for current decile
        decile_pop = 0.
        decile_inc = 0.
        # sort by income
        income_list,pop_list = (list(t) for t in zip(*sorted(zip(income_list, pop_list))))
        for ctpop, ctinc in zip(pop_list,income_list):
            cumulative_pop += ctpop
            decile_pop += ctpop
            decile_inc += ctinc*ctpop
            if cumulative_pop >= pop_cutoffs[(current_decile - 1)]:
    #             if cumulative_pop == pop_cutoffs[9]:
    #                 print("finish reached")
                current_decile += 1
                #store in deciles
                correction = (sum_pop/10.)/(decile_pop)
                decile_inc *= correction 
                #print(correction)
                #print("Decile:",decile_pop,"full/10:",(sum_pop/10.),"with correction",decile_pop*correction)
                
                deciles[(current_decile-1)][0].append(decile_inc)
                deciles[(current_decile-1)][1].append(city_pop)
                #reset
                decile_pop = 0
                decile_inc = 0

        sum_inc = np.sum(np.array(income_list)*np.array(pop))

        
def standardModelFunc(x,y0,delta):
    return y0*(x**(1+delta))
def standardLogResidual(params, x, data):
    model = standardModelFunc(x,params['y0'],params['delta'])
    return np.log(data)-np.log(model)
        
def standardLogMod(x,logy0,delta):
    return logy0+x*(1+delta)
    
plt.figure()
ax = plt.gca()
decileFits = {}
decileModels = {}
decileParams = {}
for i in range(1,11):
    x= list(deciles[i][1])
    y = list(deciles[i][0])
    #print("decile:",i,"x,y = ",x,y)
    plt.scatter(x,y,s=2)
    decileParams[i] = lmfit.Parameters()
    decileParams[i].add_many(('y0',8*list(y)[0]/list(x)[0],True,0,1),('delta',0.,True))
    

    decileModels[i] = lmfit.Model(standardModelFunc)
    decileFits[i] = lmfit.minimize(standardLogResidual, decileParams[i], args=(x, y))
    decileParams[i] = decileFits[i].params
    
    #using curve_fit
    logx = np.log(x)
    logy = np.log(y)
    coeff, var = curve_fit(standardLogMod, logx, logy)
    
    lbl = "decile "+str(i)+", Beta = "+str(round((1.+coeff[1]),3))
    print(coeff)
    fity = np.exp(np.array(coeff[0],dtype=np.float128))*x**(1+coeff[1])
    #print(x,fity)
    ax.plot(x, fity,'--',label=lbl,markersize=1)
    #ax.plot(x,decileModels[i].eval(decileParams[i],x=x),'--',label=lbl)
    

ax.set_yscale("log")
ax.set_xscale("log")
plt.legend()


    
#plt.plot(np.log(deciles[1][1]),np.log([deciles[1][0]]))
```
